```python
tc = TrainConfig()
model = vgg16(4)
crit = torch.nn.CrossEntropyLoss()
opt = torch.optim.Adam(model.parameters(),tc.LEARNING_RATE)
```

****DATA****

```python
from torchvision import datasets, transforms
from torch.utils.data import random_split, DataLoader
```

```python
# dataDir = 'drive/MyDrive/datasets/small_food'
dataDir = 'D:/Projects/data/small_food'
# !ls drive/MyDrive/datasets/small_food
dataTransforms = transforms.Compose([
    transforms.Resize(size=(224,224)),
    transforms.ToTensor()
])
allData = datasets.ImageFolder(dataDir, transform=dataTransforms)
totalDataSize = len(allData)
trainSize = int(totalDataSize * 0.8)
valSize = totalDataSize - trainSize

torch.manual_seed(42)
trainData, valData = random_split(allData, [trainSize, valSize])
batchSize = 32
numWorker = 8
trainLoader = DataLoader(trainData, batchSize, shuffle = True, num_workers=numWorker)
valLoader = DataLoader(valData, batchSize, shuffle = True, num_workers=numWorker)
```
